```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request
import requests
```

```python
url = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
import ssl

req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
uClient = urlopen(req, context=ctx)
page_html = uClient.read()
uClient.close()
```

```python
data_cols = ['Product Name', 'Sale & Actual_Price', 'Review & Ratings',
                 'Camera','Offer', 'Memory', 'Battery', 'Display'] 
```

```python
data = {}.fromkeys(data_cols, [])
```

```python
data
```

```python
data = {'Product Name': [],
       'Sale & Actual_Price': [],
       'Review & Ratings': [],
       'Camera':[],
       'Offer': [],
       'Memory': [], 
       'Battery': [],
       'Display': [],
       'Rating_number': []}

for contain in containers:
    # Product name
    pn_tag = contain.find('div', class_=product_class_name)
    product_name = pn_tag.text.strip() if pn_tag else None
    data['Product Name'].append(product_name)

    # Reviews & rating text and rating number
    rr_tag = contain.find('div', class_='_5OesEi')  # Updated class name
    reviews_and_ratings = rr_tag.text.strip() if rr_tag else None
    data['Review & Ratings'].append(reviews_and_ratings)

    # Rating number (stars)
    rating_tag = contain.find('div', class_='_3LWZlK')  # Updated class name for rating
    rating_number = rating_tag.text.strip() if rating_tag else None
    data['Rating_number'].append(rating_number)

    # Specifications: memory, display, camera, battery
    specifications = contain.find('div', class_='_6NESgJ')
    if specifications:
        specs_li = specifications.find_all('li')
        data['Memory'].append(specs_li[0].text.strip() if len(specs_li) > 0 else None)
        data['Display'].append(specs_li[1].text.strip() if len(specs_li) > 1 else None)
        data['Camera'].append(specs_li[2].text.strip() if len(specs_li) > 2 else None)
        data['Battery'].append(specs_li[3].text.strip() if len(specs_li) > 3 else None)
    else:
        data['Memory'].append(None)
        data['Display'].append(None)
        data['Camera'].append(None)
        data['Battery'].append(None)

    # Price
    price_tag = contain.find('div', class_='hl05eU')
    price = price_tag.text.strip() if price_tag else None
    data['Sale & Actual_Price'].append(price)

    # Offer
    offer_tag = contain.find('div', class_='UkUFwK')
    if offer_tag:
        data['Offer'].append(offer_tag.text.strip())
    else:
        data['Offer'].append('No Offer')
```
